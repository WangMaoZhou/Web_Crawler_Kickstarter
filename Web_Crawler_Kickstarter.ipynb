```python
from bs4 import BeautifulSoup as bs
import requests
import json
import pandas as pd
```

### 1.類別主頁面爬取所有商品資訊

```python
state = str('live')#產品狀態
category_id = str('16')#類別編號
sort = str('most_funded')#排序方式
```

```python
url="https://www.kickstarter.com/discover/advanced?state="+state+"&category_id="+category_id+"&sort="+sort+"&seed=2584289"
htmlfile = requests.get(url).text
soup = bs(htmlfile,'lxml')
```

```python
#計算頁數
div_a = soup.find('b',class_= "count ksr-green-500")
for p in div_a:
    p
a = []
a = p.strip().split(" ")
ttl_articles = int(int(a[0]) / 12)
print(ttl_articles)
```

```python
item=[]
itemurl=[]#產品頁面網址
for i in range(1,2):
    urls = url+"&page="+str(i)
    #print(urls)
    htmlfile = requests.get(urls).text
    soup = bs(htmlfile,'lxml')
    div_i = soup.find_all('div', {'class': 'js-react-proj-card grid-col-12 grid-col-6-sm grid-col-4-lg'})#.text
    ab = []
    for dataproject in div_i:
        ab.append(dataproject.get('data-project')) 
    len(ab)
    for j in range(0,len(ab)) :
        wt = json.loads(ab[j]) 
        item.append([wt['id'],wt["name"],wt['goal'],wt['pledged']])
        print(i,wt['id'],wt["name"],wt['goal'],wt['pledged'])
        itemurl.append(wt["urls"]["web"]["project"])#產品頁面網址
        print(i,wt["urls"]["web"]["project"])
```

```python
df1 = pd.DataFrame(item,columns =['id','name','goal','pledged'])
df1
```

```python
df1.to_csv('kickstarter_item.csv',encoding='utf_8_sig')
```

### 2.商品頁面爬取商品資訊

```python
item = []
for i in range(0,len(itemurl)):
    url = itemurl[i]
    print(url)
    htmlfile = requests.get(url).text
    soup = bs(htmlfile,'lxml')
    div_ii = soup.find('div',class_= "bg-grey-100")
    data = div_ii.get('data-initial')
    wt = json.loads(data)
    ws = wt["project"]
    item.append([ws["name"],ws["goal"]["amount"],ws["pledged"]["amount"],ws["percentFunded"],ws["goal"]["currency"],
                 ws["backersCount"],ws["commentsCount"],ws["location"]["displayableName"]])
    print(ws["name"],ws["goal"]["amount"],ws["pledged"]["amount"],ws["percentFunded"],ws["goal"]["currency"],
          ws["backersCount"],ws["commentsCount"],ws["location"]["displayableName"])
```

```python
df2 = pd.DataFrame(item,columns =['name','goal','pledged','percentFunded','currency','backersCount','commentsCount','location'])
```
